```octave
L = 20;
C = 20;
l0 = 10;
c0 = 10;
alpha = 4;
beta = 2;
lig = 1:L;
col = 1:C;
sigma = 1;
%Noise
b = sigma*randn(20);
%Sky 
s = 6.514;
%Amplitdue star
amp = 12.5425450;

p_th = [s;amp; l0;c0;alpha;beta]

[Col,Lig] = meshgrid(col,lig);
nu = [l0; c0; alpha; beta];
Gal = Moffat(nu,Lig,Col);

D = s + amp*Gal + b;
d = D(:);
gal=Gal(:);

figure
imagesc(D);colorbar
%%


```
